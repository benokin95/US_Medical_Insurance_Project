# U.S. Medical Insurance Costs

```python
import csv
import pandas as pd

ages = []
sexes = []
bmis = []
num_kids = []
is_smoker = []
regions = []
charges = []
insurance_dict = dict()

with open ('insurance.csv', newline='') as patient_info:
    reader = csv.DictReader(patient_info)
    for row in reader:
        ages.append(int(row['age']))
        sexes.append(row['sex'])
        bmis.append(float(row['bmi']))
        num_kids.append(int(row['children']))
        is_smoker.append(row['smoker'])
        regions.append(row['region'])
        charges.append(round(float(row['charges']), 2))
        ids = [n for n in range(1, len(ages)+1)]
        idx_pointer = 0
        for id in ids:
            insurance_dict[id] = {'Age': ages[idx_pointer],
                                 'Sex': sexes[idx_pointer],
                                 'BMI': bmis[idx_pointer],
                                 'Num Kids': num_kids[idx_pointer],
                                 'Is Smoker': is_smoker[idx_pointer],
                                  'Region': regions[idx_pointer],
                                 'Insurance Charge': charges[idx_pointer]}
            idx_pointer += 1
```

```python
def avg_by_num_kids(dict):
    total_0, total_1, total_2, total_3, total_4, total_5 = 0, 0, 0, 0, 0, 0
    count_0, count_1, count_2, count_3, count_4, count_5 = 0, 0, 0, 0, 0, 0
    for id in ids:
        if dict[id]['Num Kids'] == 0:
            total_0 += dict[id]['Insurance Charge']
            count_0 += 1
        elif dict[id]['Num Kids'] == 1:
            total_1 += dict[id]['Insurance Charge']
            count_1 += 1
        elif dict[id]['Num Kids'] == 2:
            total_2 += dict[id]['Insurance Charge']
            count_2 += 1
        elif dict[id]['Num Kids'] == 3:
            total_3 += dict[id]['Insurance Charge']
            count_3 += 1
        elif dict[id]['Num Kids'] == 4:
            total_4 += dict[id]['Insurance Charge']
            count_4 += 1
        else:
            total_5 += dict[id]['Insurance Charge']
            count_5 += 1
    avg_0 = total_0 / count_0
    avg_1 = total_1 / count_1
    avg_2 = total_2 / count_2
    avg_3 = total_3 / count_3
    avg_4 = total_4 / count_4
    avg_5 = total_5 / count_5
    print(f'The average cost for a patient with 0 children is {round(avg_0, 2)} dollars.')
    print(f'The average cost for a patient with 1 child is {round(avg_1, 2)} dollars.')
    print(f'The average cost for a patient with 2 children is {round(avg_2, 2)} dollars.')
    print(f'The average cost for a patient with 3 children is {round(avg_3, 2)} dollars.')
    print(f'The average cost for a patient with 4 children is {round(avg_4, 2)} dollars.')
    print(f'The average cost for a patient with 5 children is {round(avg_5, 2)} dollars.')
          
            
    
        
```

```python
def avg_by_age(dict):
    total_b1, total_b2, total_b3, total_b4, total_b5 = 0, 0, 0, 0, 0
    count_b1, count_b2, count_b3, count_b4, count_b5 = 0, 0, 0, 0, 0
    for id in ids:
        if dict[id]['Age'] >= 18 and dict[id]['Age'] < 26:
            total_b1 += dict[id]['Insurance Charge']
            count_b1 += 1
        elif dict[id]['Age'] >= 26 and dict[id]['Age'] < 36:
            total_b2 += dict[id]['Insurance Charge']
            count_b2 += 1
        elif dict[id]['Age'] >= 36 and dict[id]['Age'] < 46:
            total_b3 += dict[id]['Insurance Charge']
            count_b3 += 1
        elif dict[id]['Age'] >= 46 and dict[id]['Age'] < 56:
            total_b4 += dict[id]['Insurance Charge']
            count_b4 += 1
        else:
            total_b5 += dict[id]['Insurance Charge']
            count_b5 += 1
    avg_b1 = total_b1 / count_b1
    avg_b2 = total_b2 / count_b2
    avg_b3 = total_b3 / count_b3
    avg_b4 = total_b4 / count_b4
    avg_b5 = total_b5 / count_b5
    print(f'The average cost for a patient whose age is between 18 and 25 is ${round(avg_b1, 2)}.')
    print(f'The average cost for a patient whose age is between 26 and 35 is ${round(avg_b2, 2)}.')
    print(f'The average cost for a patient whose age is between 36 and 45 is ${round(avg_b3, 2)}.')
    print(f'The average cost for a patient whose age is between 46 and 55 is ${round(avg_b4, 2)}.')
    print(f'The average cost for a patient whose age is between 56 and 64 is ${round(avg_b2, 2)}.')
```

```python
avg_by_age(insurance_dict)
```

```python
def avg_by_sex(dict):
    total_m, total_f = 0, 0
    count_m, count_f = 0, 0
    for id in ids:
        if dict[id]['Sex'] == 'male':
            total_m += dict[id]['Insurance Charge']
            count_m += 1
        else:
            total_f += dict[id]['Insurance Charge']
            count_f += 1
    avg_m = total_m / count_m
    avg_f = total_f / count_f
    print(f'The average insurance cost for a male patient is ${round(avg_m, 2)}.')
    print(f'The average insurance cost for a female patient is ${round(avg_f, 2)}.')
```

```python
avg_by_sex(insurance_dict)
```
